```python
import plotly.express as px

# Creating an interactive scatter plot for Open, High, Low, and Close prices of Bitcoin
fig = px.scatter(data_btc, x='Date', y=['Open', 'High', 'Low', 'Close'],
                 labels={'value': 'Price (USD)', 'variable': 'Price Type'},
                 title='Bitcoin (BTC) Price: Open, High, Low, Close')

# Setting different colors for each price type
fig.update_traces(marker=dict(size=5),
                  selector=dict(type='scatter', mode='markers'))

# Enhancing interactivity
fig.update_layout(hovermode='closest')
fig.show()
```

## Percentage Increase in the Price 

```python
# Parsing the 'Date' column correctly and grouping data by Year-Month
data_btc['Date'] = pd.to_datetime(data_btc['Date'])
data_btc['YearMonth'] = data_btc['Date'].dt.to_period('M')
monthly_data = data_btc.groupby('YearMonth').agg({'Open': 'first', 'Close': 'last'})

# Calculating the monthly percentage increase
monthly_data['Percentage Increase'] = ((monthly_data['Close'] - monthly_data['Open']) / monthly_data['Open']) * 100

# Finding the month with the highest percentage increase
max_increase_month = monthly_data['Percentage Increase'].idxmax()
max_increase_value = monthly_data['Percentage Increase'].max()

# max_increase_month, max_increase_value
print(f"The monthly data for BitCoin shows that the month with the highest percentage increase was {max_increase_month} with a value of {max_increase_value}")
```

```python
# next task , we have to find which month had the most percentage of change in the price of the BitCoin, Ethereum and Cardano
# we have to tell the month which had highest change in price and the percentage of change of bitCoin, Ethereum and Cardano

# for that, we will look at the percentage change of the price of the BitCoin, Ethereum and Cardano in each month
# we will use the pct_change() method to calculate the percentage change of the price of the BitCoin, Ethereum and Cardano in each month
# then we will find the month which had the highest percentage change in the price of the BitCoin, Ethereum and Cardano

# we will start by calculating the percentage change of the price of the BitCoin, Ethereum and Cardano in each month
data_btc['Date'] = pd.to_datetime(data_btc['Date'])
data_btc['Month'] = data_btc['Date'].dt.month
btc_monthly_change = data_btc.groupby('Month')['Close'].pct_change()
btc_monthly_change = btc_monthly_change.dropna()
btc_monthly_change = btc_monthly_change.reset_index()
btc_monthly_change = btc_monthly_change.rename(columns={'Close': 'Percentage Change'})

data_eth['Date'] = pd.to_datetime(data_eth['Date'])
data_eth['Month'] = data_eth['Date'].dt.month
eth_monthly_change = data_eth.groupby('Month')['Close'].pct_change()
eth_monthly_change = eth_monthly_change.dropna()
eth_monthly_change = eth_monthly_change.reset_index()
eth_monthly_change = eth_monthly_change.rename(columns={'Close': 'Percentage Change'})


data_ada['Date'] = pd.to_datetime(data_ada['Date'])
data_ada['Month'] = data_ada['Date'].dt.month
ada_monthly_change = data_ada.groupby('Month')['Close'].pct_change()
ada_monthly_change = ada_monthly_change.dropna()
ada_monthly_change = ada_monthly_change.reset_index()
ada_monthly_change = ada_monthly_change.rename(columns={'Close': 'Percentage Change'})

# now we have to find the month which had the highest percentage change in the price of the BitCoin, Ethereum and Cardano
btc_monthly_change_max = btc_monthly_change.loc[btc_monthly_change['Percentage Change'].idxmax()]
eth_monthly_change_max = eth_monthly_change.loc[eth_monthly_change['Percentage Change'].idxmax()]
ada_monthly_change_max = ada_monthly_change.loc[ada_monthly_change['Percentage Change'].idxmax()]

print('BitCoin Month with Highest Percentage Change:', btc_monthly_change_max['Month'], 'with', round(btc_monthly_change_max['Percentage Change']*100, 2), '% change')
print('Ethereum Month with Highest Percentage Change:', eth_monthly_change_max['Month'], 'with', round(eth_monthly_change_max['Percentage Change']*100, 2), '% change')
print('Cardano Month with Highest Percentage Change:', ada_monthly_change_max['Month'], 'with', round(ada_monthly_change_max['Percentage Change']*100, 2), '% change')
```
