```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
```

## Scraping URLs of All Products

```python
no_of_pages = 105

product_urls = []

for page in range(no_of_pages):
    url = "https://modivo.pl/c/damskie-kurtki-i-plaszcze?p=" + str(page+1)
    try:
        driver = webdriver.Chrome()
        # Create a new Chrome driver instance
        driver.get(url)

        # Wait for the page to load (you might need to adjust the waiting time)
        driver.implicitly_wait(5)

        # Get the page source after it's loaded
        html = driver.page_source

    except Exception as e:
        print(e)

    # Parse the HTML using BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    lis = soup.find_all('a', class_='text-link product-card-link')
    
    for a in lis:
        product_urls.append(a['href'])

driver.quit()

products_df = pd.DataFrame(product_urls, columns=['Product Links'])
len(products_df)
```

```python
# Saving product links to csv
products_df.to_csv('modivo_product_pages_urls.csv', index=False)
```

## Scraping Images From Each Product

```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import TimeoutException

products_df = pd.read_csv('modivo_product_pages_urls.csv')

images_urls = []
for i, url in enumerate(list(products_df['Product Links'][0:749])):
    try:
        driver = webdriver.Chrome()
        
        full_url = "https://modivo.pl" + url
        
        # driver.get(full_url)

        driver.set_page_load_timeout(4)  # Set page load timeout to 4 seconds

        try:
            driver.get(full_url)
        except TimeoutException:
            pass  # Handle timeout exception, page will not load fully

        html = driver.page_source

    except Exception as e:
        print(e)

    soup = BeautifulSoup(html, 'html.parser')

    divs = soup.find_all('div', class_='media-item')

    for j, div in enumerate(divs):
        img = div.find('img')
        if img and img.has_attr('src'):
            jpg_url = img['src']
            images_urls.append([i+1, j+1, jpg_url.replace('product', 'zoomapp')])
            print(f"Fetched {i+1} product {j+1} image url.")
    driver.quit()

images_df = pd.DataFrame(images_urls, columns=['Product', 'Image No#', 'Product URL'])
images_df.to_csv('modivo_images_urls_1.csv', index=False)

images_df.head(20)
```

## Downloading Images

```python
# now downloading the images
import os
os.makedirs('modivo_images', exist_ok=True)

for index, row in images_df.iterrows():
    prod = row['Product']
    img = row['Image No#']
    if prod <= 379:
        continue
    elif prod >500:
        break
    else:
        print(f"Downloading {prod} product {img} image.")
        with open(f'modivo_images/{str(prod)}_{str(img)}.jpg', 'wb') as f:
            f.write(requests.get(row['Product URL']).content)
```

```python
# Resizing the images to 768 x 1024 using imutils
import imutils
import cv2
os.makedirs('modivo_images_resized', exist_ok=True)

for index, row in images_df.iterrows():
    if row['Product'] <= 736:
        continue
    else:
        try:
            print(f"Resizing {row['Product']} product {row['Image No#']} image.")
            img = cv2.imread(f'modivo_images/{str(row["Product"])}_{str(row["Image No#"])}.jpg')
            resized = imutils.resize(img, width=768, height=1024)
            # resize using cv2
            # resized = cv2.resize(img, (768, 1024), interpolation=cv2.INTER_AREA)
            cv2.imwrite(f'modivo_images_resized/{str(row["Product"])}_{str(row["Image No#"])}.jpg', resized)
        except Exception as e:
            print("Error: ", e)
```
