For male: Generation of product URLs


```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

no_of_pages = 35  # Set the number of pages you want to scrape
base_url = "https://www.zalando.co.uk/mens-clothing-jackets/?p="

product_urls = []

# Initialize the Chrome driver
driver = webdriver.Chrome()

# Iterate over each page
for page in range(1, no_of_pages + 1):
    url = f"{base_url}{page}"
    try:
        # Load the page
        driver.get(url)
        
        # Wait for the page to load (you might need to adjust the waiting time)
        driver.implicitly_wait(20)

        # Get the page source after it's loaded
        html = driver.page_source

        # Parse the HTML using BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')

        # Find all product links
        links = soup.find_all('a', class_='_LM JT3_zV CKDt_l CKDt_l LyRfpJ')
        
        # Append the product URLs to the list
        for link in links:
            product_urls.append(link['href'])

    except Exception as e:
        print(e)

# Quit the driver after scraping
driver.quit()

# Create a DataFrame of the product URLs
products_df = pd.DataFrame(product_urls, columns=['Product Links'])
print(len(products_df))
```
